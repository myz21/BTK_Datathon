Üniversite bölümü, lise bölümü verilerinde benzersiz değerlerin fazlalığının aşırı öğrenme (overfitting) yapabileceğini düşündüğüm için bunları da çıkarıyorum:

```python
df_train.drop(columns=["Bölüm","Lise Bolumu"], inplace=True)
```

```python
df_train.nunique()
```

ID ve değerlendirme puanı dışındaki değerleri çıkarıp hangi benzersiz değerlere sagip olduğumuzu kontrol edelim: 

```python
# Hariç tutmak istediğiniz sütunlar
excluded_columns = ['Degerlendirme Puani', 'id']

# Tüm sütun isimlerini al
columns = df_train.columns

# Hariç tutulacak sütunlar dışında kalan sütunları seç
columns_to_check = [col for col in columns if col not in excluded_columns]

# Her bir sütun için benzersiz değerleri hesapla ve yazdır
for column in columns_to_check:
    unique_values = df_train[column].unique()
    print(column)
    print(unique_values)
    print()  # Satır boşluğu
```
