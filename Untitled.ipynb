    return next((c for c in cities if c.lower() in city), None)

import re

def extract_numbers(text):
    if isinstance(text, str):
        # Sayıları bul ve en büyüğünü döndür
        numbers = re.findall(r'\d+', text)
        return int(max(numbers, key=int)) if numbers else 0
    return 0

    'Dogum Yeri Düzeltilmis',

    'Bölüm_encoded'
]

# Encode each column
for column in columns_to_encode:
    # Ensure column is treated as string
    df_train[column] = df_train[column].astype(str)

    # Initialize LabelEncoder
    le = LabelEncoder()

    # Fit and transform the column
    df_train[column] = le.fit_transform(df_train[column])

    # Save the encoder for this column
    label_encoders[column] = le
